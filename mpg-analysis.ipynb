```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

Create a visual showing the relationship between horsepower and fuel economy.

```python
sns.scatterplot(data=mpg, x="horsepower", y="mpg")
plt.xlabel("Horsepower")
plt.ylabel("Miles Per Gallon (MPG)")
plt.title("Relationship Between Horsepower and Fuel Economy")
plt.show()
```

There is a negative correlation between horsepower and fuel economy. As horsepower increases, mpg decreases, indicating that more powerful cars tend to be less fuel-efficient.

Create a table summarizing fuel economy by country of origin


```python
mpg_summary = mpg.groupby("origin")["mpg"].agg(["mean", "median", "min", "max"]).round(2)
mpg_summary
```

Vehicles from Japan and Europe tend to have higher fuel economy (higher mpg) compared to those from the USA. Where as 
American cars, on average, have lower mpg, possibly due to their larger engine sizes and higher horsepower.

Create a visual showing how fuel economy has changed over time


```python
sns.lineplot(data=mpg, x="model_year", y="mpg", errorbar=None, marker="o")
plt.xlabel("Model Year")
plt.ylabel("Miles per Gallon (MPG)")
plt.title("Fuel Economy Change Over Time")
plt.show()
```

Fuel economy has improved over the years, especially after the 1970s, likely due to increasing fuel efficiency regulations and advances in automotive technology. The trend shows a gradual increase in mpg over time, indicating better energy efficiency in newer vehicles.

Create two other visuals or tables related to fuel economy.

```python
sns.histplot(mpg["mpg"], bins=20, kde=True)
plt.xlabel("Miles per Gallon (MPG)")
plt.ylabel("Count")
plt.title("Distribution of Fuel Economy (MPG)")
plt.show()
```

The distribution is right-skewed, meaning most vehicles have moderate fuel economy, but there are a few very fuel-efficient cars with high mpg. The peak suggests that the majority of vehicles fall within a mid-range mpg category.

```python
sns.boxplot(data=mpg, x="cylinders", y="mpg")
plt.xlabel("Number of Cylinders")
plt.ylabel("Miles per Gallon (MPG)")
plt.title("Fuel Economy by Cylinder Count")
plt.show()
```

Vehicles with fewer cylinders (4 or 6) tend to have higher fuel economy. Cars with 8 cylinders have significantly lower mpg, reinforcing the idea that larger engines consume more fuel.
